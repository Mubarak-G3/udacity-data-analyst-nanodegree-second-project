## Analysis on WeRateDogs Twitter Data

## Introduction

The Twitter account called WeRateDogs provides ratings for people's pets in exchange for a witty remark about the dog. Most of the time, ten is used as the denominator in these ratings. The numeratorts on the other hand, everytime the value is greater than ten, as shown by the numbers 11/10, as well as 12/10, 13\10, and so on. WeRateDogs has  gained extensive international media coverage as a result of its more than 4 million followers. Because of Twitter's developer API, it is feasible to access information about the canines who have posted on the social media site. It is possible to combine data from a neural network system that is capable of predicting different dog breeds with Twitter  JSON data to generate a highly effective combination. This huge amount of data may be acquired, analyzed, cleaned, and integrated with other data using data wrangling techniques to provide insight into the data's relationship with other data.

#### Dog Name Frequency
The  cleaned  and  consolidated  data  collection  consists  of  around  850  tweets,  which  include, among other things, the names of the dogs in question. There are also predictions for the various dog breeds included in the data. In order to make the predictions, the tweets were fed through a neural network that can distinguish between different dog breeds. The golden retriever is the most frequently predicted dog breed in the first forecast for a specific dog, followed by the Labrador retriever  in the second and third predictions.
